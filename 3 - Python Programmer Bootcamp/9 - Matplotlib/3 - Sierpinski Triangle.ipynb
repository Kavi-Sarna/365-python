# Sierpinski Triangle

First Transformation
\begin{equation}
x_{n+1} = 0.5x_n
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}
Second Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 0.5
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n + 0.5
\end{equation}
Third Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 1
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}

```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
```

```python
plt.rc('figure', figsize=(16, 16))
```

```python
def transformation_1(p):
  x = 0.5 * p[0]
  y = 0.5 * p[1]
  return x, y

def transformation_2(p):
  x = 0.5 * p[0] + 0.5
  y = 0.5 * p[1] + 0.5
  return x, y

def transformation_3(p):
  x = 0.5 * p[0] + 1
  y = 0.5 * p[1]
  return x, y


transformations = [transformation_1, transformation_2, transformation_3]
x_cords = [0]
y_cords = [0]
x, y = 0, 0

for i in range(1000000):
  trans = choice(transformations)
  x, y = trans((x, y))
  x_cords.append(x)
  y_cords.append(y)
```

```python
plt.plot(x_cords, y_cords, 'o')
plt.title('Sierpinski Triangle')
plt.savefig('Sierpinski.png')
```
