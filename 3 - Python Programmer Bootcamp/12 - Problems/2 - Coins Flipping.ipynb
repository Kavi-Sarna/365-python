```python
n = 1001
coins = [0]*n
```

```python
for i in range(1, n):
  for j in range(0, n, i):
    coins[j] = 1 - coins[j]
```

```python
print(coins)
```

```python
d = {}

for i, v in enumerate(coins):
  if v != 0:
    d[i] = v
```

```python
d
```

```python
L = []

for k, v in d.items():
  L.append(k)
```

```python
print(L)
```

```python
import math

L2 = [math.sqrt(num) for num in L]

print(L2)
```
