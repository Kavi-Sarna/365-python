```python
def add(L):
  ''' Adds the integer items of a list'''

  size = len(L)
  total = 0
  iterator = 0

  while iterator < size:
    total = total + L[iterator]
  return total
```

```python
my_list = [1, 2, 3, 4, 5, 6, 7]
add(my_list)
```
