# Matplotlib Random Walk project

### First we import the necessary libraries

```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
```

### Then we change the default display settings for our plots

```python
plt.rc('figure', figsize=(12, 6))
```

### Now let's create a Random Walk

```python
step_option = [1, -1]
step_choice = choice(step_option)
step_choice
```

```python
walk = []
walk.append(choice(step_option))

for step in range(1, 1000):
  walk.append(walk[step-1] + choice(step_option))

print(walk)
```

```python
plt.plot(walk)
plt.show()
```

### Using functions would be better

```python
def rand_walk(step_num):
  walk = []
  walk.append(choice([1, -1]))

  for step in range(1, step_num+1):
    walk.append(walk[step-1] + choice([1, -1]))

  return walk
```

```python
random_walk = rand_walk(100)
print(random_walk)
```

```python
def plot_walk(walk):
  plt.plot(walk)
  plt.title('Random Walk')
  plt.xlabel('No of Steps')
  plt.ylabel('Distance from Origin')
  plt.show()
```

```python
plot_walk(rand_walk(1000))
```

### We can create a function to plot more than 1 random walk

```python
def plot_walk(num_of_walks, num_of_steps):
  label_list = list(range(num_of_walks))

  for i in range(num_of_walks):
    plt.plot(rand_walk(num_of_steps), label = f'Plot Number {label_list[i]}')
    plt.xlabel('Number of Steps')
    plt.ylabel('Distance from Origin')
    plt.title('Random Walks')
    plt.legend(loc='lower left')

  plt.show()
```

```python
plot_walk(5, 100000)
```
