```python
from PIL import Image, ImageFilter, ImageFont, ImageDraw
from matplotlib.pyplot import imshow
```

```python
file_name = 'hillside.jpg'
im = Image.open(file_name)
imshow(im)
```

```python
print(im.size)
print(im.format_description)
print(im.mode)
```

```python
crop = (200, 200, 600, 600)
cropped_image = im.crop(crop)
imshow(cropped_image)
```

```python
resized_image = im.resize((400, 200))
imshow(resized_image)
```

```python
grey_scale = im.convert('LA')
imshow(grey_scale)
```

```python
effect = im.filter(ImageFilter.GaussianBlur(radius = 10))
imshow(effect)
```

```python
draw = ImageDraw.Draw(im)
font = ImageFont.truetype("arial.ttf", 200)
draw.text((100, 100), "Hillside Photo!", font=font)
imshow(im)
```
